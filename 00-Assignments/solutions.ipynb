```python
# 1/1
import gurobipy as gp
from gurobipy import GRB

# Create a new model
m = gp.Model("mip1")

# Create variables

x1 = m.addVar(name="x1")
x2 = m.addVar(name="x2")

m.setObjective(5 * x1 +  x2, GRB.MAXIMIZE)

# Add constraints

m.addConstr(3 * x1 - 2 * x2 <= 10, "c0")
m.addConstr(x1 + 2 * x2 <= 8, "c1")
m.addConstr(x1 + x2 <= 5, "c2")

# Optimize model

m.optimize()

# Print the optimal values of the variables


for v in m.getVars():
    print(f"{v.varName}: {v.x}")
# Print the optimal objective value
print(f"Obj: {m.objVal}")
```
